```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import yfinance as yf
import statsmodels.api as sm
```

```python
data = pd.read_csv('FAANG.csv',parse_dates=True)
data.index = pd.to_datetime(data.Date)
data.drop('Date',axis=1,inplace=True)
nasdaq = yf.Ticker('^IXIC').history(period='max').Close
data['NASDAQ'] = nasdaq
data['NASDAQ_Ret'] = nasdaq.pct_change()
data
```

```python
ff = pd.read_csv('FF_Daily.csv')
ff.index=pd.to_datetime(ff.Date, format='%Y%m%d', errors='ignore')
ff.drop(['Date','Mkt-RF'],axis=1,inplace=True)
ff
```

```python
data = pd.concat([data,ff],axis=1)
data.dropna(inplace=True)
data.drop(data.columns[0:5],axis=1,inplace=True)
data.drop('NASDAQ',axis=1,inplace=True)
data
```

### French Fama

```python
stockRet = data['FB_Ret']-data['RF']
stockRet.name = 'stockRet'
marketRet = data['NASDAQ_Ret']-data['RF']
marketRet.name = 'marketRet'
SMB = data['SMB']
SMB.name='SMB'
HML = data['HML']
HML.name='HML'
regressData= pd.concat([stockRet,marketRet,SMB,HML],axis=1)
regressData.dropna(inplace=True)
regressData
```

```python
def famafrench(stock_name):
    #Prepare DataFrame
    stockRet = data[stock_name]-data['RF']
    stockRet.name = 'stockRet'
    marketRet = data['NASDAQ_Ret']-data['RF']
    marketRet.name = 'marketRet'
    SMB = data['SMB']
    SMB.name='SMB'
    HML = data['HML']
    HML.name='HML'
    regressData= pd.concat([stockRet,marketRet,SMB,HML],axis=1)
    regressData.dropna(inplace=True)
    stock = stock_name.replace('_Ret','')
    #Regress
    regress = sm.OLS(regressData['stockRet'],sm.add_constant(regressData.iloc[:,1:]))
    result = regress.fit()
    print('%s: Î±=%.6f, Î²=%.6f, SMB coef.=%.6f, HML coef.=%.6f'
          %(stock,result.params[0],result.params[1],result.params[2],result.params[3]))
    print('Corresponding p-value: Î±=%.6f, Î²=%.6f, SMB coef.=%.6f, HML coef.=%.6f'
         %(result.pvalues[0],result.pvalues[1],result.pvalues[2],result.pvalues[3]))
```

```python
famafrench('FB_Ret')
```
